```python
import tensorflow as tf
#from keras.models import load_model
```

```python
model = tf.keras.models.load_model('modelfull.h5')
```

```python
test_dir = r'F:\\GAIT RECOGNITION\\fulldataset\\test'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

height=240
width=240
batch_size=32


# Test generator
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(test_dir, 
                                                  target_size=(height,width), 
                                                  batch_size=batch_size,
                                                  color_mode='grayscale',
                                                  shuffle = False,
                                                  class_mode='categorical')
```

```python
import os

os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'
```

```python
test_acc = model.evaluate(test_generator,verbose=1)                  
print(test_acc)
```

```python
preds = model.predict(test_generator)
```

```python
import numpy as np

train_dir = r'F:\\GAIT RECOGNITION\\90degreegei\\train'


# Training generator
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(train_dir, 
                                                    target_size=(height,width),
                                                    batch_size=batch_size,
                                                    color_mode='grayscale',
                                                    class_mode='categorical')



predicted_class_indices=np.argmax(preds,axis=1)
print(predicted_class_indices)


labels = (train_generator.class_indices)
labels = dict((v,k) for k,v in labels.items())
predictions = [labels[k] for k in predicted_class_indices]
```

```python
import pandas as pd


filenames=test_generator.filenames
print(filenames)
results=pd.DataFrame({"Filename":filenames,"Predictions":predictions})
results.to_csv("results.csv",index=False)
```

```python
from PIL import Image
from skimage import transform

def load(filename):
   np_image = Image.open(filename)
   np_image = np.array(np_image).astype('float32')/255
   np_image = transform.resize(np_image, (240, 240, 1))
   np_image = np.expand_dims(np_image, axis=0)
   return np_image


toBePred = load(r'F:\\GAIT RECOGNITION\\fulldataset\\test\\080\\080-nm-06-000.png')

single_pred = model.predict(toBePred) 

#print(single_pred)

predicted_class_indices=np.argmax(single_pred,axis=1)
print(predicted_class_indices)
```
